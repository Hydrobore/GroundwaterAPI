```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the site name as a variable
site_name = "Carlton Lake Bottom"

# Load the combined dataframe from the CSV file
combined_df = pd.read_csv('ConcatCSVdata.csv')

# Convert the 'date' column to datetime format
combined_df['date'] = pd.to_datetime(combined_df['date'])

# Define the keywords to plot based on the site name
keywords_to_plot = [f"{site_name} logged", f"{site_name}"]

# Filter the dataframe to include only the selected keywords
filtered_df = combined_df[combined_df['keyword'].isin(keywords_to_plot)]

# Sort the filtered dataframe by 'date'
filtered_df = filtered_df.sort_values(by='date')

# Define hex string colors for each keyword
color_map_hex = {
    f"{site_name} logged": "#E69F00",    # Orange
    f"{site_name}": "#000000",           # Black


}

# Set the figure size
plt.figure(figsize=(12, 8))

# Plot each keyword's data with lines and smaller markers
for keyword in keywords_to_plot:
    keyword_data = filtered_df[filtered_df['keyword'] == keyword]
    plt.plot(keyword_data['date'], keyword_data['value'], marker='o', linestyle='-',
             label=keyword, markersize=1, color=color_map_hex[keyword])

# Set plot title and labels using the site name
plt.title(f'Groundwater Levels for {site_name} Multi-level borehole monitoring')
plt.xlabel('Date')
plt.ylabel('Groundwater Level, mAOD')
plt.legend()
plt.grid(True)

# Save the plot using the site name as the filename
plt.savefig(f'{site_name}.png')

# Show the plot
plt.show()
```
